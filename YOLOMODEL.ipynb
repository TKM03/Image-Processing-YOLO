```python
# Save the trained model as a .pt file for future use
from ultralytics import YOLO
import cv2

model_path = 'trained_yolov8_model.pt'
model = YOLO(model_path)
```

```python
# Path to the image you want to test
image_path = 'test5.jpg'

# Run the prediction
results = model.predict(source=image_path, save=False)

# Load the original image using OpenCV
image = cv2.imread(image_path)

# Check if the image was loaded successfully
if image is None:
    print("Error: Could not load the image.")
else:
    # Loop over results and draw the bounding boxes
    for result in results:
        for box in result.boxes:
            # Extract the bounding box coordinates
            x1, y1, x2, y2 = map(int, box.xyxy[0])  # Convert to integer
            conf = box.conf.item()  # Get confidence score
            class_id = box.cls.item()  # Get class index

            # Draw the bounding box on the image
            cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)

            # Add a label with the class and confidence score
            label = f'{result.names[class_id]}: {conf:.2f}'
            cv2.putText(image, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # Display the image with bounding boxes using OpenCV
    cv2.imshow('YOLOv8 Test Result', image)
    cv2.waitKey(0)  # Wait for a key press
    cv2.destroyAllWindows()  # Close the OpenCV window
```

```python
# Save the image with bounding boxes to a file
output_image_path = 'output_test_result.jpg'
cv2.imwrite(output_image_path, image)
print(f"Output saved to {output_image_path}")
```

```python
# Path to your trained YOLOv8 model
model_path = 'trained_yolov8_model.pt'

# Load the YOLOv8 model
model = YOLO(model_path)

# Open the webcam (0 is the default camera; you can change the index for external cameras)
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("Error: Could not open the webcam.")
    exit()

while True:
    # Capture frame-by-frame from the webcam
    ret, frame = cap.read()

    if not ret:
        print("Failed to grab frame.")
        break

    # Run the YOLO model on the captured frame
    results = model.predict(source=frame, save=False)

    # Loop over results and draw bounding boxes
    for result in results:
        for box in result.boxes:
            # Extract the bounding box coordinates
            x1, y1, x2, y2 = map(int, box.xyxy[0])  # Convert to integer
            conf = box.conf.item()  # Get confidence score
            class_id = box.cls.item()  # Get class index

            # Draw the bounding box on the frame
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)

            # Add a label with the class and confidence score
            label = f'{result.names[class_id]}: {conf:.2f}'
            cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # Display the frame with bounding boxes in a window
    cv2.imshow('YOLOv8 Real-Time Detection', frame)

    # Break the loop when 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam and close all OpenCV windows
cap.release()
cv2.destroyAllWindows()
```
